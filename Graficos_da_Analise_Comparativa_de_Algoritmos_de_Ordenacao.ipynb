```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Gera os valores de x
n = np.linspace(0.01, 20, 400)

# Complexidades Teóricas
O_n2 = n**2
O_nlogn = n * np.log2(n)
O_n = n
O_logn = np.log2(n)
O_1 = n / n
O_0 = n * 0
```

## Complexidade X Tamanho

```python
# Plota as funções
plt.plot(n, O_n2, label='Selection = O(n²)', color='red')
plt.plot(n, O_nlogn, label='Quick, Merge, Intro = O(n ⋅ log(n))', color='yellow')
plt.plot(n, O_n, label='Insertion, Bubble = O(n)', color='green')

# Personalização
plt.title('Complexidade X Qnt. de Elementos (Melhor caso)')
plt.xlabel('Quantidade de Elementos no Vetor (n)')
plt.ylabel('Complexidade')
plt.axhline(0, color='black', linewidth=0.5)  # linha horizontal y=0
plt.axvline(0, color='black', linewidth=0.5)  # linha vertical x=0
plt.grid(True)
plt.legend()

# Define limites dos eixos (somente valores positivos)
plt.xlim(0, 20)           # Eixo x de 0 até 20
plt.ylim(0, 400)          # Eixo y de 0 até 400

# Mostra o gráfico
plt.show()
```

```python
# Plota as funções
plt.plot(n, O_n2, label='Selection, Insertion, Bubble = O(n²)', color='red')
plt.plot(n, O_nlogn, label='Quick, Merge, Intro = O(n ⋅ log(n))', color='yellow')

# Personalização
plt.title('Complexidade X Qnt. de Elementos (Caso médio)')
plt.xlabel('Quantidade de Elementos no Vetor (n)')
plt.ylabel('Complexidade')
plt.axhline(0, color='black', linewidth=0.5)  # linha horizontal y=0
plt.axvline(0, color='black', linewidth=0.5)  # linha vertical x=0
plt.grid(True)
plt.legend()

# Define limites dos eixos (somente valores positivos)
plt.xlim(0, 20)           # Eixo x de 0 até 20
plt.ylim(0, 400)          # Eixo y de 0 até 400

# Mostra o gráfico
plt.show()
```
